# 1. Removing locations with duplicate links.

One of the ways to remove duplication is to identify duplicates by their links. However, Google maps tacks on additional information to the links. This information can change as the same page is visited during different scrapes. Meaning that the links and different and cannot be used to identify duplicate entries of the same location.

To fix this we are going to normalize the links, stripping away any additional unecessary information.

```python
import re
from urllib.parse import urlsplit, urlunsplit, parse_qsl, urlencode, unquote
 
import pandas as pd
 
# --- patterns for the stable identifiers ------------------------------------
_FTID = re.compile(r'(?:!1s|[?&]ftid=)(0x[0-9a-fA-F]+:0x[0-9a-fA-F]+)')
_FTID_ANY = re.compile(r'(0x[0-9a-fA-F]+:0x[0-9a-fA-F]+)')
_CID = re.compile(r'[?&]cid=(\d+)')
_PLACEID = re.compile(r'(?:place_id[:=]|query_place_id=)?(ChIJ[0-9A-Za-z_\-]{8,})')
_MID = re.compile(r'!16s(%2[fF][gm]%2[fF][0-9A-Za-z_]+|/[gm]/[0-9A-Za-z_]+)')
 
# query params that never identify the place — always discarded
_JUNK_PARAMS = {
    'entry', 'g_ep', 'g_st', 'hl', 'gl', 'ved', 'sa', 'source', 'shorturl',
    'authuser', 'ucbcb', 'output', 'gs_lcp', 'ictx', 'cshid',
}
 
 
def _extract_identifier(url):
    """Return (kind, value) for the most reliable place identifier, or None."""
    m = _FTID.search(url)
    if m:
        return ('ftid', m.group(1).lower())
 
    m = _CID.search(url)
    if m:
        return ('cid', m.group(1))
 
    m = _PLACEID.search(url)
    if m:
        return ('place_id', m.group(1))
 
    m = _MID.search(url)
    if m:
        return ('mid', unquote(m.group(1)))
 
    # last resort: a bare hex feature id sitting anywhere in the string
    m = _FTID_ANY.search(url)
    if m:
        return ('ftid', m.group(1).lower())
 
    return None
 
 
def _strip_noise(url):
    """Best-effort cleanup when no stable id is found.
 
    Drops the fragment, the `@lat,lng,zoom` viewport path segment, and all
    non-identifying query params; keeps the place path so distinct places stay
    distinct.
    """
    parts = urlsplit(url)
 
    # remove the @viewport segment(s) from the path
    segments = [s for s in parts.path.split('/') if not s.startswith('@')]
    path = '/'.join(segments)
 
    # keep only query params that could identify the place
    kept = [(k, v) for k, v in parse_qsl(parts.query, keep_blank_values=False)
            if k.lower() not in _JUNK_PARAMS]
    query = urlencode(kept)
 
    return urlunsplit((parts.scheme.lower(), parts.netloc.lower(),
                       path.rstrip('/'), query, ''))
 
 
def standardize_gmaps_url(url, mode='url'):
    """Standardize a single Google Maps link.
 
    Parameters
    ----------
    url : str
        The raw link. Non-strings / NaN / empty values are returned unchanged.
    mode : {'url', 'key'}
        'url' -> a clean canonical Google Maps URL (default).
        'key' -> just the bare identifier (tightest possible match key).
 
    Returns
    -------
    str
        Canonical link (or key). Same place -> same output.
    """
    if not isinstance(url, str) or not url.strip():
        return url
 
    url = url.strip()
 
    # Short links (maps.app.goo.gl / goo.gl/maps) can't be expanded offline.
    # Two identical short links still match; just trim noise.
    if 'goo.gl' in url or 'maps.app' in url:
        cleaned = _strip_noise(url)
        return cleaned if mode == 'url' else cleaned
 
    ident = _extract_identifier(url)
 
    if ident is None:
        cleaned = _strip_noise(url)
        return cleaned if mode == 'url' else cleaned
 
    kind, value = ident
 
    if mode == 'key':
        return f'{kind}:{value}'
 
    # canonical, deterministic URL forms
    if kind == 'cid':
        return f'https://www.google.com/maps?cid={value}'
    if kind == 'place_id':
        return f'https://www.google.com/maps/place/?q=place_id:{value}'
    if kind == 'ftid':
        return f'https://www.google.com/maps?ftid={value}'
    if kind == 'mid':
        return f'https://www.google.com/maps?q={value}'
 
    return _strip_noise(url)  # unreachable, but safe
 
 
def standardize_url_column(df, col='google_maps_url',
                           keep_original=True, mode='url'):
    """Standardize an entire URL column in a DataFrame.
 
    Parameters
    ----------
    df : pd.DataFrame
    col : str
        Name of the link column.
    keep_original : bool
        If True, the raw links are preserved in `{col}_raw` before overwriting.
    mode : {'url', 'key'}
        Passed through to `standardize_gmaps_url`.
    """
    df = df.copy()
    if keep_original:
        df[f'{col}_raw'] = df[col]
    df[col] = df[col].map(lambda u: standardize_gmaps_url(u, mode=mode))
    return df
 
 
if __name__ == '__main__':
    # Two captures of the SAME place: different viewport + query junk.
    a = ('https://www.google.com/maps/place/Riyadh+Park/@24.7553,46.6296,17z/'
         'data=!3m1!4b1!4m6!3m5!1s0x3e2efef1c1234567:0x89abcdef01234567!8m2'
         '!3d24.7553!4d46.6296!16s%2Fg%2F11abc?entry=ttu&g_ep=EgoyMDI0')
    b = ('https://www.google.com/maps/place/Riyadh+Park/@24.7551,46.6299,18z/'
         'data=!4m6!3m5!1s0x3e2efef1c1234567:0x89abcdef01234567!8m2!3d24.755'
         '!4d46.629?entry=ttu&hl=ar')
    # A genuinely different place.
    c = 'https://maps.google.com/?cid=11112222333344445555'
 
    for raw in (a, b, c):
        print(standardize_gmaps_url(raw))
 
    assert standardize_gmaps_url(a) == standardize_gmaps_url(b), 'should match!'
    assert standardize_gmaps_url(a) != standardize_gmaps_url(c), 'should differ!'
    print('\nOK: duplicates collapse, distinct places stay distinct.')
```

```python
df_standardized = standardize_url_column(df)
```

Now that the links are standardized, we can use them to remove locations that have been scraped twice.

```python
def _as_missing(df, columns):
    """Return a copy where empty / whitespace-only strings become NaN.
 
    Only touches object (text) columns; numeric columns already use NaN.
    """
    cleaned = df.copy()
    for col in columns:
        if col in cleaned and cleaned[col].dtype == object:
            cleaned[col] = cleaned[col].replace(r'^\s*$', np.nan, regex=True)
    return cleaned


def add_completion_score(df):
    """Add a `completion_score` column counting non-null attributes per row.
    """
    df = df.copy()
    source = _as_missing(df, df.columns)
    df['completion_score'] = source[df.columns].notna().sum(axis=1)
    return df

def drop_duplicate_links(df, link_col='google_maps_url', score_col='completion_score'):
    """Remove rows that share an identical link, keeping the most complete one.
 
    Rows with a missing/empty link are kept as-is (they aren't true duplicates
    of each other). Original row order is preserved.
    """
    link = df[link_col].astype('string').str.strip()
    has_link = link.notna() & (link != '')
 
    with_link = df[has_link]
    without_link = df[~has_link]
 
    deduped = (
        with_link
        .sort_values(score_col, ascending=False, kind='stable')  # best first
        .drop_duplicates(subset=link_col, keep='first')
    )
 
    result = pd.concat([deduped, without_link]).sort_index()
    return result
 
 
def clean(df):
    """Run both steps and print a short summary."""
    df = add_completion_score(df)
 
    before = len(df)
    df = drop_duplicate_links(df)
    removed = before - len(df)
 
    print(f"Rows before : {before}")
    print(f"Rows after  : {len(df)}")
    print(f"Removed      : {removed} exact-link duplicate(s)")
    return df
```

## 3. Phone Number Cleaning and Standardization

```python
def clean_phone(phone):
    if pd.isna(phone):
        return pd.Series([phone, "Unknown"])

    phone = str(phone).strip() # convert to str to use Regex and remove spaces at beginning and end.
    digits = re.sub(r"\D", "", phone) # if symbols and spaces del

    if digits == "":
        return pd.Series([phone, "Unknown"])

    phone_type = "Unknown"

    # Unified Number
    if digits.startswith("9200"):
        phone_type = "Unified"

    elif digits.startswith("800"):
        phone_type = "Unified"

    # Mobile
    elif digits.startswith("05"):
        digits = "966" + digits[1:]
        phone_type = "Mobile"

    elif digits.startswith("5") and len(digits) == 9:
        digits = "966" + digits
        phone_type = "Mobile"

    elif digits.startswith("9665"):
        phone_type = "Mobile"

    # Landline
    elif digits.startswith("0"):
        digits = "966" + digits[1:]
        phone_type = "Landline"

    elif digits.startswith("966"):
        if len(digits) >= 4 and digits[3] == "5":
            phone_type = "Mobile"
        else:
            phone_type = "Landline"

    return pd.Series([digits, phone_type])

df[["phone_number", "phone_type"]] = df["phone_number"].apply(clean_phone)

# Move phone_type next to phone_number
cols = df_deduplicated.columns.tolist()
phone_idx = cols.index("phone_number")
cols.remove("phone_type")
cols.insert(phone_idx + 1, "phone_type")
df = df[cols]

df_deduplicated.to_csv("saihat_google_maps_deduplicated.csv", index=False)
```

## 4. Website URL Cleaning and Standardization

```python
def clean_website(url):
    if pd.isna(url):
        return np.nan
    
    url = str(url).strip()
    url = url.replace(" ", "") # del spaces

    # if empty
    if url == "":
        return np.nan

    # Standardize website URLs to use HTTPS
    if url.startswith("http://"):
        url = "https://" + url[len("http://"):]

    elif url.startswith("www."):
        url = "https://" + url

    elif not url.startswith("https://"):
        url = "https://" + url

    return url

df["website"] = df["website"].apply(clean_website)
df_deduplicated.to_csv("saihat_google_maps_deduplicated.csv", index=False)
```
